Create 3 random matrices and compute their inverses, if they exist

```python
## Imports
import numpy as np
```

```python
## Code for matrix A
matrix = [[1, 2, 3], [4, 5, -4], [-3, -2, -1]]

matrix_np = np.array(matrix)

print("For matrix A: ")
print(matrix_np)
try:
    inverse = np.linalg.inv(matrix_np)
    print("Inverse found:")
    print(inverse)

    identity = np.round((inverse @ matrix_np), 2)
    print("Proof of inverse: ")
    print(identity)
except np.linalg.LinAlgError:
    print("No inverse found")
```
