```python
import numpy as np
import pandas as pd
```

## part (a)

```python
df = pd.read_csv("kc_house_data.csv")

df = df.drop(columns=["id", "date", "zipcode"])

print(df.info())
```

```python
lowest_mean, highest_mean = [float('inf'), ""], [-float('inf'), ""]
lowest_var, highest_var = [float('inf'), ""], [-float('inf'), ""]


for feature in df:
    data = df[feature]

    if pd.api.types.is_numeric_dtype(data):
        # Calculate statistics
        avg = data.mean()
        var = data.var()
        min_val = data.min()
        max_val = data.max()
        
        # Print the results
        print(f"Statistics for {feature}: ")
        print(f"  Average: {avg}")
        print(f"  Variance: {var}")
        print(f"  Minimum: {min_val}")
        print(f"  Maximum: {max_val}")

        if avg < lowest_mean[0]:
            lowest_mean = [avg, feature]
        if avg > highest_mean[0]:
            highest_mean = [avg, feature]
        if var < lowest_var[0]:
            lowest_var = [var, feature]
        if var > highest_var[0]:
            highest_var = [var, feature]


print(f"Lowest Mean: {lowest_mean[1]}: {lowest_mean[0]}")
print(f"Highest Mean: {highest_mean[1]}: {highest_mean[0]}")
print(f"Lowest Variance: {lowest_var[1]}: {lowest_var[0]}")
print(f"Highest Variance: {highest_var[1]}: {highest_var[0]}")
```

## part (b)

```python
## Compute correlation
corrs = df.corr()

price_corrs = corrs["price"]

print("Correlation Coeffcients with Price: ")
print(price_corrs)
```
